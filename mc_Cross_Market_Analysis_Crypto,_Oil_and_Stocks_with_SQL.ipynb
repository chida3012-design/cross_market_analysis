# **Join Queries**

```python
mc_cursor.execute("use cross_market_analysis;")
mc_cursor.execute("""SELECT
    year(c.date),
    avg(c.price) AS BitcoinAvgPrice,
    avg(o.price_inr) AS OilAvgPrice,
    (avg(c.price) / avg(o.price_inr)) AS BTC_to_Oil_Ratio
FROM
    cryptoprices c
JOIN
    oil_prices o ON year(c.date) = year(o.date)
WHERE
    year(c.date) = '2025' and c.coin_id='bitcoin'
group by year(c.date),year(o.date)""")
query_21=mc_cursor.fetchall()
query_21
```

```python
mc_cursor.execute("use cross_market_analysis;")
mc_cursor.execute("""SELECT
    c.date,
    c.price AS BTC_price,
    s.close AS SP_Close,
    ((c.price - LAG(c.price) OVER (ORDER BY c.date)) / LAG(c.price) OVER (ORDER BY c.date)) * 100 AS BTC_Pct_Change,
    ((s.close - LAG(s.close) OVER (ORDER BY s.date)) / LAG(s.close) OVER (ORDER BY s.date)) * 100 AS SP_Pct_Change,
    CASE
        WHEN ((c.price - LAG(c.price) OVER (ORDER BY c.date)) *
              (s.close - LAG(s.close) OVER (ORDER BY s.date))) > 0
        THEN 'Same Direction'
        ELSE 'Opposite'
    END AS Correlation_Sign
FROM
    cryptoprices c
JOIN
    stock_prices s ON c.date = s.date
WHERE c.coin_id='bitcoin' and s.ticker='GSPC'
ORDER BY
    c.date DESC;""")
query_22=mc_cursor.fetchall()
query_22
```

```python
mc_cursor.execute("use cross_market_analysis;")
mc_cursor.execute("""SELECT
    c.date,
    c.price AS ethereum_price,
    s.close AS IXIC_close_price,
    c.price / s.close AS ethereum_IXIC_Ratio
FROM
    cryptoprices c
JOIN
    stock_prices s ON c.date = s.date
WHERE year(c.date) = '2025' and
    c.coin_id='ethereum' and
    s.ticker='IXIC';""")
query_23=mc_cursor.fetchall()
query_23
```

```python
mc_cursor.execute("use cross_market_analysis;")
mc_cursor.execute("""SELECT
    o.date,
    o.price_inr AS oil_price,
    ((o.price_inr - prev_o.price_inr) / prev_o.price_inr) * 100 AS oil_percent_change,
    c.price AS btc_price,
    ((c.price - prev_c.price) / prev_c.price) * 100 AS btc_percent_change
FROM
    oil_prices o
JOIN
    oil_prices prev_o ON o.date = DATE_ADD(prev_o.date, INTERVAL 1 DAY)
JOIN
    cryptoprices c ON o.date = c.date
JOIN
   cryptoprices prev_c ON c.date = DATE_ADD(prev_c.date, INTERVAL 1 DAY)
WHERE
     c.coin_id='bitcoin' and ((o.price_inr - prev_o.price_inr) / prev_o.price_inr) > 0.02
ORDER BY
    o.date DESC;""")
query_24=mc_cursor.fetchall()
query_24
```

```python
mc_cursor.execute("use cross_market_analysis;")
mc_cursor.execute("""WITH DailyCryptoChanges AS (
    SELECT
        c.date,
        c.coin_id,
        c.price,
        ((c.price - LAG(c.price) OVER (PARTITION BY c.coin_id ORDER BY c.date))
        / LAG(c.price) OVER (PARTITION BY c.coin_id ORDER BY c.date)) * 100 AS coin_pct_change
    FROM cryptoprices c
    WHERE c.coin_id IN ('bitcoin', 'ethereum', 'tether')
),
DailyNiftyChange AS (
    SELECT
        s.date,
        s.close AS nifty_close,
        ((s.close - LAG(s.close) OVER (ORDER BY s.date))
        / LAG(s.close) OVER (ORDER BY s.date)) * 100 AS nifty_pct_change
    FROM stock_prices s where s.ticker='NSEI'
)

SELECT
    d.date,
    d.coin_id,
    ROUND(d.coin_pct_change, 2) AS coin_daily_pct,
    ROUND(n.nifty_pct_change, 2) AS nifty_daily_pct,
    CASE
        WHEN d.coin_pct_change > n.nifty_pct_change THEN 'Outperforming'
        WHEN d.coin_pct_change < n.nifty_pct_change THEN 'Underperforming'
        ELSE 'Neutral'
    END AS trend_comparison
FROM DailyCryptoChanges d
JOIN DailyNiftyChange n ON d.date = n.date
ORDER BY d.date DESC, d.coin_id;""")
query_25=mc_cursor.fetchall()
query_25
```
